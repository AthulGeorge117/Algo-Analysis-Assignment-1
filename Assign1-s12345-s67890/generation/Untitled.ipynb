```python
from matplotlib import pyplot as plt
```

```python
listyy = [1.1034, 1.0436, 1.0028, 1.9915,10 ]
linked_list = [23,45,78,98,45]
trie = [23,45,78,98,45]

listyy = [ x/5 for x in listyy]
linked_list = [ x/5 for x in linked_list]
trie = [ x/5 for x in trie]

dataSizes = [500, 1000, 5000, 10000, 50000, 100000, 200000]

plt.plot(dataSizes , listyy , "r-")
plt.plot(dataSizes , linked_list , "b-")
plt.plot(dataSizes , trie , "g-")
plt.xlabel("dataSizes")
plt.ylabel("Execution Time in ms")
plt.title("Add_Word_Frequency")
plt.show()
```
